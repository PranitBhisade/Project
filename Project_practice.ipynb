```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,scale
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("G://project.csv")
```

```python
data.Impact = data['Impact'].astype('str')
data.Urgency = data['Urgency'].astype('str')
```

```python
data[data.Impact.str.contains('NS') == True]
data[data.Urgency.str.contains('NA') == True]
```

```python
data.drop(data.index[[38305]],inplace=True)
```

```python
data = data.replace('NS',np.nan)
data = data.replace('NA',np.nan)
```

```python
data.head()
```

```python
data = data.fillna(method = 'ffill')
```

```python
data.head()
```

```python
data.Impact = data['Impact'].astype('int')
```

```python
data.Urgency = data['Urgency'].astype('int')
```

```python
enc = LabelEncoder()
```

```python
data.Impact = enc.fit_transform(data.Impact)
data.Urgency = enc.fit_transform(data.Urgency)
#data.No_of_Related_Incidents = enc.fit_transform(data.No_of_Related_Incidents)
data.Priority = enc.fit_transform(data.Priority)

```

```python
X = data.loc[:,['Impact','No_of_Related_Incidents']]
Y = data.Priority
```

```python
#Importing KNN model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors= 3)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.3 ,random_state = 15)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_predict = model.predict(X_test)
```

```python
Y_predict
```

```python
#Importing Preprocessing
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(Y_test,Y_predict),classification_report(Y_test,Y_predict))
```

```python
print(accuracy_score(Y_test,Y_predict))
```
